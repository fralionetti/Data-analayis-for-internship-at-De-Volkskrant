Noordpool dataset updated until 2024

```python
#importing package
import pandas as pd

#reading CSV file and skipping rows
df = pd.read_csv("C:/Users/lionet000/Downloads/vishop_extent (3).csv", skiprows=3)

#Selecting relevant columns
df_subset = df[['#date', '1980\'s Average', '1990\'s Average', '2000\'s Average', '2010\'s Average', '2019', '2020', '2021', '2022', '2023', '2024']]

#Renaming columns
df_subset.rename(columns={
    "1980's Average": "Gemiddelde 1980-89",
    "1990's Average": "Gemiddelde 1990-99",
    "2000's Average": "Gemiddelde 2000-2009",
    "2010's Average": "Gemiddelde 2010-2019"
}, inplace=True)

#Converting '#date' to datetime and makling a new month-day column
df_subset['#date'] = pd.to_datetime(df_subset['#date'], format='%Y-%m-%d')
df_subset['maand_dag'] = df_subset['#date'].dt.strftime('%m-%d')

#Dropping the date column
df_subset.drop('#date', axis=1, inplace=True)

#Rounding numeric columns
df_subset = df_subset.applymap(lambda x: round(x / 1000000, 3) if pd.api.types.is_numeric_dtype(x) else x)

#Saving to CSV
df_subset.to_csv('Noordpool tot 2024.csv', index=False)

#Display the new DataFrame
print(df_subset)
```
