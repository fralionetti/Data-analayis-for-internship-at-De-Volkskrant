# Step 1: I open the first batch of the datasets and merge them together

Initially, I only open the datasets from 2016 to 2022 as the 2015 dataset does not have the "Foreign" column.

```python
#importing the necessary package
import pandas as pd

#creating a list of file paths
file_paths = [
    "C:/Users/lionet000/Documents/migration EU/eurodac 2020/Hits category 1 vs category 1  - EURODAC 2020.xlsx",
    "C:/Users/lionet000\Documents/migration EU/eurodac 2019/Hits category 1 vs category 1 - EURODAC 2019.xlsx",
    "C:/Users/lionet000/Documents/migration EU/eurodac 2021/Hits category 1 vs category 1  - EURODAC 2021.xlsx", 
    "C:/Users/lionet000/Documents/migration EU/eurodac 2022/Hits category 1 vs category 1  - EURODAC 2022.xlsx",
    "C:/Users/lionet000/Documents/migration EU/eurodac 2018/Hits category 1 vs category 1  - EURODAC 2018.xlsx",
    "C:/Users/lionet000/Documents/migration EU/eurodac 2017/Hits category 1 vs category 1  - EURODAC 2017.xlsx",
    "C:/Users/lionet000/Documents/migration EU/eurodac 2016/Hits category 1 vs category 1  - EURODAC 2016.xlsx"]

#creating a dictionary where to store the files
dfs = {}

#Extracting the file name without extension
for file_path in file_paths:
    file_name = file_path.split("/")[-1].split(".")[0]
    #Reading the Excel file into a pandas DataFrame
    dfs[file_name] = pd.read_excel(file_path)

#Printing the datasets
for file_name, df in dfs.items():
    print(f"\n{file_name}:")
    print(df.head())
```

# Step 2: I add the year column and merge these datasets together

```python
#I shorten the name of the datasets, add the year column and print them to check that everything went well

EURODAC1_2020 = 'Hits category 1 vs category 1  - EURODAC 2020'

# Check if the target dataset is in the dictionary
if EURODAC1_2020 in dfs:
    # Access the target dataset (DataFrame)
    EURODAC1_2020 = dfs[EURODAC1_2020]
    EURODAC1_2020['Year'] = '2020'
    print(EURODAC1_2020.head())
    
    
    
EURODAC1_2019 = 'Hits category 1 vs category 1 - EURODAC 2019'

# Check if the target dataset is in the dictionary
if EURODAC1_2019 in dfs:
    # Access the target dataset (DataFrame)
    EURODAC1_2019 = dfs[EURODAC1_2019]
    EURODAC1_2019['Year'] = '2019'
    print(EURODAC1_2019.head())
    
    
EURODAC1_2018 = 'Hits category 1 vs category 1  - EURODAC 2018'

# Check if the target dataset is in the dictionary
if EURODAC1_2018 in dfs:
    # Access the target dataset (DataFrame)
    EURODAC1_2018 = dfs[EURODAC1_2018]
    EURODAC1_2018['Year'] = '2018'
    print(EURODAC1_2018.head())


EURODAC1_2021 = 'Hits category 1 vs category 1  - EURODAC 2021'

# Check if the target dataset is in the dictionary
if EURODAC1_2021 in dfs:
    # Access the target dataset (DataFrame)
    EURODAC1_2021 = dfs[EURODAC1_2021]
    EURODAC1_2021['Year'] = '2021'
    print(EURODAC1_2021.head())
    
EURODAC1_2022 = 'Hits category 1 vs category 1  - EURODAC 2022'

# Check if the target dataset is in the dictionary
if EURODAC1_2022 in dfs:
    # Access the target dataset (DataFrame)
    EURODAC1_2022 = dfs[EURODAC1_2022]
    EURODAC1_2022['Year'] = '2022'
    print(EURODAC1_2022.head())
    
EURODAC1_2017 = 'Hits category 1 vs category 1  - EURODAC 2017'

# Check if the target dataset is in the dictionary
if EURODAC1_2017 in dfs:
    # Access the target dataset (DataFrame)
    EURODAC1_2017 = dfs[EURODAC1_2017]
    EURODAC1_2017['Year'] = '2017'
    print(EURODAC1_2017.head())
    
EURODAC1_2016 = 'Hits category 1 vs category 1  - EURODAC 2016'

# Check if the target dataset is in the dictionary
if EURODAC1_2016 in dfs:
    # Access the target dataset (DataFrame)
    EURODAC1_2016 = dfs[EURODAC1_2016]
    EURODAC1_2016['Year'] = '2016'
    print(EURODAC1_2016.head())
```

```python
#merging the datasets together
new_df_migration = pd.concat([EURODAC1_2018, EURODAC1_2019, EURODAC1_2020, EURODAC1_2021, EURODAC1_2022, EURODAC1_2017, EURODAC1_2016], ignore_index=False)
print(new_df_migration)
```

# Step 3: I open the 2015 dataset and I merge it together with the rest

This dataset does not have a foreign column, so I need to create one.

```python
#Opening the dataset
file_path2 = "C:/Users/lionet000/Documents/migration EU/eurodac 2015/Hits category 1 vs category 1  - EURODAC 2015.xlsx"
EURODAC1_2015 = pd.read_excel(file_path2)
print(EURODAC1_2015.head())
```

```python
#Creating the year and the Foreign colun
EURODAC1_2015['Year'] = '2015'
EURODAC1_2015['Foreign'] = EURODAC1_2015['Total'] - EURODAC1_2015['Local']
print(EURODAC1_2015.head())
```

```python
EURODAC1_df = pd.concat([EURODAC1_2015, new_df_migration], ignore_index=True)
print(EURODAC1_df.head())
```

# Step 4: cleaning the merged dataset

```python
#The only think that I need to do to clean the dataset is to trnasform the NAs values into 0
EURODAC1_final = EURODAC1_df.fillna(0)

#I check that everything went well and save the fial dataset
print(EURODAC1_final)
EURODAC1_final.to_csv("EURODAC1_final", index=False)
```

# Step 5: I analyse the first 2 RQs. Through which coutries do migrants go when entering Europe and what about those arriving in the Netherlands?

RQ 1: Through which coutries do migrats go when entering Europe? As I want to look at which migrats go through once entering Europe, I am interested in looking at how many hits where reported in each country. The more hits in acountry, the more popular the destination is for migrants either as a place of arrival or as a place to go through. Therefore, I want to look at the the total hits per each country registered in each year.

RQ2: Which route did the migrants arriving in the Netherlands take? I want to look at the how many hits were reported in each EU country for the fingerprints take in the Netherland in a given year.

For the sake of time, I create one dataset with information for both RQs. Hence, I keep the the Total row and the Netherland row for each year

First, I create a dataset with all of the information over all of the years. In this wasy, I will be able to anlyse how the situation evolved over the years.

```python
#I speficy the 2 rows that I want to keep and the column thta I want to exclude as they are not relevant for this analsyis. 
desired_values = ['NL', 'Total']
columns_to_exclude = ['Local', 'Foreign', 'Total']
migrantsroute_df_final = EURODAC1_final[EURODAC1_final['Member states'].isin(desired_values)].drop(columns=columns_to_exclude)

#I print and save the fial dataset
print(migrantsroute_df_final.head())
migrantsroute_df_final.to_csv("migrationroute_df_final.csv", index=False)
```

Second, I create a sub-dataset for each year. In this way, I will be able to look at the data year by year later on.

```python
columns_to_exclude = ['Year']

# Convert the 'Year' column to a string to ensure proper comparison
migrantsroute_df_final['Year'] = migrantsroute_df_final['Year'].astype(str)

# Create a dictionary of DataFrames, where each key is a unique year
yearly_subsets = {}
for year in migrantsroute_df_final['Year'].unique():
    yearly_subsets[year] = migrantsroute_df_final[migrantsroute_df_final['Year'] == year].drop(columns=columns_to_exclude)

# Print and save each dataset
for year, subset in yearly_subsets.items():
    # Print the dataset
    print(f"Migration route {year}:\n{subset}\n")

    # Save the dataset to a CSV file (adjust the file path as needed)
    subset.to_csv(f"Migration route_{year}.csv", index=False)
```
